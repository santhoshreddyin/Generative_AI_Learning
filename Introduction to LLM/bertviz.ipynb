```python
from transformers import AutoModel, AutoTokenizer, utils

utils.logging.set_verbosity_error()  # Suppress standard warnings

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModel.from_pretrained("gpt2", output_attentions=True)
```

```python
from bertviz import head_view

inputs = tokenizer.encode(
    "Give me test cases for customer creation flow on a digital bank", return_tensors="pt"
)
outputs = model(inputs)
attention = outputs[-1]
tokens = tokenizer.convert_ids_to_tokens(inputs[0])

head_view(attention, tokens, layer=7, heads=[10])
```
